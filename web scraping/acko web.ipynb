```python
result = soup.find_all('div', {'data-testid':'listingcardesktop'})
```

```python
# Get all <p> tags with the specified class
elements = result[0].find_all('p', {'class': "styles__ParaWithoutMargins-sc-a6403e05-34 iAPjDz"})
elements[5].get_text()
```

```python
for el in elements:
    text = el.get_text()
    if "Available in" in text:
        print(text)
```

```python
import re

for el in elements:
    text = el.get_text()
    match = re.search(r"Available in (\d+) colors?", text)
    if match:
        num_colors = int(match.group(1))
        print(f"Number of colors available: {num_colors}")
```
